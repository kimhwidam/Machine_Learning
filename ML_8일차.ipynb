### Naive Bayes 분류

먼저 영어로

```python
# Naive Bayes는 지도학습이라서 정답을 알려주어야 함
from nltk.tokenize import word_tokenize
import nltk

train = [('i like you', 'pos'), ('i hate you', 'neg'),
         ('you like me', 'neg'), ('i like her', 'pos')]

# 전체 말뭉치를 만들어 주기
all_words = set(
    word.lower() for sentence in train for word in word_tokenize(sentence[0])
)
all_words
```

말 뭉치에서 각 단어 유무 파악

1. train 에서 x(=문장)를 하나씩 가져오기
2. x[0](첫 문장의 첫 번째 것)을 띄어쓰기로 분리
3. all_words에 있는 모든 단어(word)를 가지고 2.에 있는지 확인

-> 말뭉치에 대비해서 단어가 있고 없음을 표기

```python
t = [({word: (word in word_tokenize(x[0])) for word in all_words}, x[1]) for x in train]
t
```

```python
# Naive Bayes 훈련 시작
classifier = nltk.NaiveBayesClassifier.train(t)
# 가장 많은 정보를 담고 있는 특성을 나열
classifier.show_most_informative_features()
```

* like가 있을 때 Positive할 확률이 1.7:1.0
* 각 단어별로 독립적으로 확률을 계산하기 때문에 naive하다

```python
# 테스트
test_sentence = 'i like MeRui'
test_sent_features = {
    word.lower(): (word in word_tokenize(test_sentence.lower())) for word in all_words
}
test_sent_features
```

```python
classifier.classify(test_sent_features)
```

결과는 positive - 감성분석

이제 한글로

```python
from konlpy.tag import Okt

pos_tagger = Okt()
# 정답을 알고 있는 문장으로 훈련용 데이터 생성
train = [
    ('메리가 좋아', 'pos'),
    ('고양이도 좋아', 'pos'),
    ('난 수업이 지루해', 'neg'),
    ('메리는 이쁜 고양이야', 'pos'),
    ('난 마치고 메리랑 놀거야', 'pos')
]

# 전체 말뭉치 만들기
all_words = set(word for sentence in train for word in word_tokenize(sentence[0]))
all_words
```

메리가, 메리는, 메리랑을 모두 다른 단어로 인식...

```python
t = [({word: (word in word_tokenize(x[0])) for word in all_words}, x[1]) for x in train]
t
```

```python
# Naive Bayes 훈련 시작
classifier = nltk.NaiveBayesClassifier.train(t)
classifier.show_most_informative_features()
```

```python
# 테스트
test_sentence = '난 수업이 마치면 메리랑 놀거야'
test_sent_features = {
    word.lower(): (word in word_tokenize(test_sentence.lower())) for word in all_words
}
test_sent_features
```

```python
classifier.classify(test_sent_features)
```

* 부정으로 나옴...
* 한글은 형태소 분석이 필수
* 형태소 분석이 필요함

```python
# 형태소 분석을 한 후 품사를 단어 뒤에 붙이는 함수
def tokenize(doc):
    return ["/".join(t) for t in pos_tagger.pos(doc, norm=True, stem=True)]
```

```python
train_docs = [(tokenize(row[0]), row[1]) for row in train]
train_docs
```

```python
# 풀어서 말뭉치 만들기
tokens = [t for d in train_docs for t in d[0]]
tokens
```

```python
def term_exists(doc):
    return {word: (word in set(doc)) for word in tokens}

train_xy = [(term_exists(d), c) for d, c in train_docs]
train_xy
```

```python
# 이제 다시 학습 후 테스트
classifier = nltk.NaiveBayesClassifier.train(train_xy)
test_sentence = [('난 수업이 마치면 메리랑 놀거야')]
test_docs = pos_tagger.pos(test_sentence[0])
test_docs
```

```python
classifier.show_most_informative_features()
```

```python
test_sent_features = {word: (word in tokens) for word in test_docs}
test_sent_features
```

```python
classifier.classify(test_sent_features)
```

이제는 긍정으로 분류됨

### Count Vectorizer
문장을 점처럼 일종의 벡터로 표현할 수 있다면 두 문장 사이의 거리를 구해서 유사한 문장 찾기가 가능하지 않을까?

```python
# 문장을 벡터로 변환하는 함수
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(min_df=1)
```

```python
# 훈련용 문장 생성
contents = ['상처받은 아이들은 너무 일찍 커버려',
            '내가 상처받은 거 아는 사람 불편해',
            '잘 사는 사람들은 좋은 사람 되기 쉬워',
            '아무 일도 아니야 괜찮아']
```

```python
# 형태소 분석
from konlpy.tag import Okt

t = Okt()
contents_tokens = [t.morphs(row) for row in contents]
contents_tokens
```

```python
# 형태소로 띄어쓰기 문장 만들기
# 리스트형을 string형으로
contents_for_vectorize = []

for content in contents_tokens:
    sentence = ''
    for word in content:
        sentence = sentence + ' ' + word

    contents_for_vectorize.append(sentence)

contents_for_vectorize
```

```python
# 변환
X = vectorizer.fit_transform(contents_for_vectorize)
X
```

```python
num_samples, num_features = X.shape
num_samples, num_features
```

말뭉치가 17개

```python
# 말뭉치 보기
vectorizer.get_feature_names_out()
```

```python
X.toarray().transpose()
```

```python
# 새로운 문장으로 테스트
new_post = ['상처 받기 싫어 괜찮아']
new_post_tokens = [t.morphs(row) for row in new_post]

new_post_for_vectorize = []

for content in new_post_tokens:
    sentence = ''
    for word in content:
        sentence = sentence + ' ' + word

    new_post_for_vectorize.append(sentence)

new_post_for_vectorize
```

```python
# 벡터화 시키기
new_post_vec = vectorizer.transform(new_post_for_vectorize)
new_post_vec.toarray()
```

```python
# 기하학적인 거리 사용 함수
import scipy as sp

def dist_raw(v1, v2):
    delta = v1 - v2
    return sp.linalg.norm(delta.toarray())
```

```python
# 거리 구하기
dist = [dist_raw(each, new_post_vec) for each in X]
dist
```

```python
# 결과
print('Best post is ', dist.index(min(dist)), ', dist = ', min(dist))
print('Test post is --> ', new_post)
print('Best dist post is --> ', contents[dist.index(min(dist))])
```

### TF-IDF Vectorizer

* 한 문서에서 많이 등장한 단어에 가중치 -> (Term Freq.)
* 전체 문서에서 많이 나타나는 단어는 중요하지 않게 -> (Inverse Document Freq.)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(min_df=1, decode_error='ignore')
```

```python
X = vectorizer.fit_transform(contents_for_vectorize)
X
```

```python
num_samples, num_features = X.shape
num_samples, num_features
```

```python
X.toarray().transpose()
```

조금 다름

```python
new_post_vec = vectorizer.transform(new_post_for_vectorize)
new_post_vec.toarray()
```

```python
# 거리를 정규화시켜서 측정하는 함수
def dist_norm(v1, v2):
    v1_normalized = v1 / sp.linalg.norm(v1.toarray())
    v2_normalized = v2 / sp.linalg.norm(v2.toarray())

    delta = v1_normalized - v2_normalized

    return sp.linalg.norm(delta.toarray())
```

```python
dist = [dist_norm(each, new_post_vec) for each in X]
dist
```

```python
# 결과
print('Best post is ', dist.index(min(dist)), ', dist = ', min(dist))
print('Test post is --> ', new_post)
print('Best dist post is --> ', contents[dist.index(min(dist))])
```

### 네이버 지식인 검색 결과에서 유사한 문장 찾기

```python
import urllib.request
import json
import datetime

# 예전 Naver API 실습 코드 가져오기
def gen_search_url(api_node, search_text, start_num, disp_num):
    base = "https://openapi.naver.com/v1/search"
    node = "/" + api_node + ".json"
    param_query = "?query=" + urllib.parse.quote(search_text)
    param_start = "&start=" + str(start_num)
    param_disp = "&display=" + str(disp_num)

    return base + node + param_query + param_start + param_disp

def get_result_onpage(url):
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id", client_id)
    request.add_header("X-Naver-Client-Secret", client_secret)
    response = urllib.request.urlopen(request)
    print("[%s] Url Request Success" % datetime.datetime.now())
    return json.loads(response.read().decode("utf-8"))

def delete_tag(input_str):
    input_str = input_str.replace("<b>", "")
    input_str = input_str.replace("</b>", "")
    return input_str

def get_description(pages):
    contents = []

    for sentences in pages['items']:
        contents.append(delete_tag(sentences['description']))

    return contents

client_id = ""
client_secret = ""

url = gen_search_url('kin', '파이썬', 10, 10)
one_result = get_result_onpage(url)
one_result
```

```python
# description만 가져오기
contents = get_description(one_result)
contents
```
