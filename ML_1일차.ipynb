## 데이터 분리
* 확보한 데이터 중에서 모델 학습에 사용하지 않고 빼둔 데이터를 가지고 모델을 테스트

```python
# 새롭게 다시 시작
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
```

```python
# 훈련용 데이터, 테스트용 데이터르 8:2 비율로 나누기
from sklearn.model_selection import train_test_split

features = iris.data[:, 2:]
labels = iris.target

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2,
                                                    stratify=labels, random_state=13)
```

```python
# 훈련, 테스트 확인
X_train.shape, X_test.shape
```

```python
# 각 품종 별로 동일 비율이 아님 -> 그럴 때 stratify 옵션 사용
np.unique(y_test, return_counts=True)
```

```python
# Scikit Learn으로 Decision Tree 구현
from sklearn.tree import DecisionTreeClassifier

# 모델을 단순화시키기 위해 max_depth 조정
iris_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
# 학습
iris_tree.fit(X_train, y_train)
```

```python
plt.figure(figsize=(12, 6))
plot_tree(iris_tree)
```

```python
# 훈련 데이터 정확도 확인
y_pred_tr = iris_tree.predict(X_train)
accuracy_score(y_train, y_pred_tr)
```

```python
# 결정 경계 확인 -> 경계면이 단순해짐 -> max_depth를 줄이는 것이 좋다는 것을 확인
plt.figure(figsize=(14,8))
plot_decision_regions(X=X_train, y=y_train, clf=iris_tree, legend=2)
plt.show()
```

```python
# 테스트 데이터 정확도 확인
y_pred_test = iris_tree.predict(X_test)
accuracy_score(y_test, y_pred_test)
```

```python
# 좀 더 자세하게 결정 경계 시각화
scatter_highlight_kwargs = {'s':150, 'label':'Test data', 'alpha':0.9}
scatter_kwargs = {'s':120, 'edgecolor':None, 'alpha':0.7}

plt.figure(figsize=(12,8))
plot_decision_regions(features, y=labels, X_highlight=X_test, clf=iris_tree, legend=2,
                      scatter_highlight_kwargs=scatter_highlight_kwargs,
                      scatter_kwargs=scatter_kwargs, contourf_kwargs={'alpha':0.2})
plt.show()
```

```python
# feature를 네 개로
features = iris.data
labels = iris.target

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2,
                                                    stratify=labels, random_state=13)

iris_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
iris_tree.fit(X_train, y_train)

plt.figure(figsize=(12, 6))
plot_tree(iris_tree)
```

```python
# 모델 사용 -> 길 가다가 주운 iris가 sepal과 petal의 length, width의 값 입력

test_data = [[4.3, 2., 1.2, 1.0]]
iris_tree.predict_proba(test_data)
```
