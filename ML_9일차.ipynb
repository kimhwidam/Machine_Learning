## 실습 1 - 영화 추천

데이터 출처 - https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/tmdb_5000_movies.csv

```python
# 데이터 불러오기
import pandas as pd
import numpy as np

movies = pd.read_csv('https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/tmdb_5000_movies.csv')
movies
```

### 데이터 전처리

```python
movies.columns
```

```python
# 필요한 컬럼만 추출
movies = movies[['id', 'title', 'genres', 'vote_average', 'vote_count', 'popularity', 'keywords', 'overview']]
movies
```

```python
movies['genres'][0]
```

genres와 keywords는 컬럼 안에 dict형으로 저장됨

genres와 keywords의 내용을 list와 dict으로 복구 필요

```python
from ast import literal_eval

movies['genres'] = movies['genres'].apply(literal_eval)
movies['keywords'] = movies['keywords'].apply(literal_eval)
movies
```

```python
# dict의 value 값을 특성으로 사용하도록 변경
movies['genres'] = movies['genres'].apply(lambda x : [y['name'] for y in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [y['name'] for y in x])
movies[['genres', 'keywords']][:2]
```

```python
# genres의 각 단어들을 하나의 문장(띄어쓰기로 구분된)으로 변환
movies['genres_literal'] = movies['genres'].apply(lambda x : (' ').join(x))
movies
```

```python
# 문자열로 변환된 genres를 CountVectorize 수행
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(min_df=0, ngram_range=(1,2))
genre_mat = count_vect.fit_transform(movies['genres_literal'])
print(genre_mat.shape)
```

문장의 유사도 측정을 코사인 유사도로 수행

```python
from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim.shape)
print(genre_sim[:2])
```

```python
# genre_sim 객체에서 높은 값 순으로 정렬
genre_sim_sorted_ind = genre_sim.argsort()[:, ::-1]
print(genre_sim_sorted_ind[:1])
```

```python
# 추천 영화를 데이터프레임으로 반환하는 함수
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    title_movie = df[df['title'] == title_name]

    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, :(top_n)]

    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)

    return df.iloc[similar_indexes]
```

```python
# 영화 대부로 테스트
similar_movies = find_sim_movie(movies, genre_sim_sorted_ind, 'The Godfather', 10)
similar_movies[['title', 'vote_average']]
```

```python
movies[['title', 'vote_average', 'vote_count']].sort_values('vote_average', ascending=False)[:10]
```

다시 보니 평점과 평점을 매긴 횟수를 보면 문제 데이터가 보임

영화 선정을 위한 가중치 선정 필요

```python
# 영화 전체 평균평점과 최소 투표 횟수를 60%지점으로 지정
C = movies['vote_average'].mean()
m = movies['vote_count'].quantile(0.6)
print('C: ', round(C, 3), 'm: ', round(m, 3))
```

```python
# 가중치가 부여된 평점을 계산하기 위한 함수
def weighted_vote_average(record):
    v = record['vote_count']
    R = record['vote_average']

    return ( (v/(v+m)) * R ) + ( (m/(m+v)) * C )
```

```python
movies['weighted_vote'] = movies.apply(weighted_vote_average, axis=1)
movies
```

```python
# 전체 데이터에서 가중치가 부여된 평점 순으로 정렬한 결과
movies[['title', 'vote_average', 'weighted_vote', 'vote_count']].sort_values('weighted_vote', ascending=False)[:10]
```

```python
# 유사 영화 찾기 함수를 변경
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    title_movie = df[df['title'] == title_name]

    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, :(top_n*2)]
    similar_indexes = similar_indexes.reshape(-1)

    similar_indexes = similar_indexes[similar_indexes != title_index]

    return df.iloc[similar_indexes].sort_values('weighted_vote', ascending=False)[:top_n]
```

```python
# 영화 대부로 다시 테스트
similar_movies = find_sim_movie(movies, genre_sim_sorted_ind, 'The Godfather', 10)
similar_movies[['title', 'vote_average', 'weighted_vote']]
```

```python
# book_tags와 tags를 merge
tags_join_df = pd.merge(book_tags, tags, left_on='tag_id', right_on='tag_id', how='inner')
tags_join_df
```

```python
# to_read 데이터 불러오기
to_read = pd.read_csv('/content/to_read.csv')
to_read
```

```python
# books 데이터의 authors로 Tfidf 수행
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), min_df=0, stop_words='english')
tfidf_matrix = tf.fit_transform(books['authors'])
tfidf_matrix
```

```python
# 유사도 측정
from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim
```

```python
# The Hobbit
titles = books['title']
indices = pd.Series(books.index, index=books['title'])
indices['The Hobbit']
```

```python
# 유사도 값 호출
cosine_sim[indices['The Hobbit']]
```

```python
# 유사도 결과를 인덱스를 가진 list형으로
cosine_sim[indices['The Hobbit']].shape
```

```python
list(enumerate(cosine_sim[indices['The Hobbit']]))[:3]
```

```python
# 가장 유사한 책의 인덱스 찾기
sim_scores = list(enumerate(cosine_sim[indices['The Hobbit']]))
sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
sim_scores[:3]
```

```python
# 작가로 본 유사 책 검색
sim_scores = sim_scores[1:11]
book_indices = [i[0] for i in sim_scores]
titles.iloc[book_indices]
```

```python
# book에 tag 포함
books_with_tags = pd.merge(books, tags_join_df, left_on='book_id', right_on='goodreads_book_id', how='inner')
books_with_tags
```

```python
# 이번에는 tag로 Tfidf
tf1 = TfidfVectorizer(analyzer='word', ngram_range=(1,2), min_df=0, stop_words='english')
tfidf_matrix1 = tf1.fit_transform(books_with_tags['tag_name'].head(10000))
cosine_sim1 = linear_kernel(tfidf_matrix1, tfidf_matrix1)
```

```python
# 추천책을 반환하는 함수
titles1 = books['title']
indices1 = pd.Series(books.index, index=books['title'])

def tags_recommendations(title):
    idx = indices1[title]
    sim_scores = list(enumerate(cosine_sim1[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    book_indices = [i[0] for i in sim_scores]
    return titles.iloc[book_indices]

tags_recommendations('The Hobbit').head(20)
```
